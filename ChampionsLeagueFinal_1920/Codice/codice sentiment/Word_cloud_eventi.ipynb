```python
import pandas as pd
match_int=pd.read_csv('C:/Users/preinstalled/Downloads/match_int.csv')
```

```python
#seleziono i tweet in ligua italiana
datiITA=match_int.loc[match_int['language'] == 'it']
datiITA
```

```python
#lista di stop words
import nltk
from nltk.corpus import stopwords
stop = stopwords.words('italian')
stop.extend(['#UCL','#ChampionsLeague','#PSGBayern','#UCLFinal','#PSGFCB'])
stop
```

```python
#creo una colonna di liste e rimuovo le stop words
datiITA["text"] =datiITA["text"].str.lower().str.split()
datiITA['text']=datiITA["text"].apply(lambda x: [item for item in x if item not in stop])
```

```python
datiITA.head()
```

```python
#slicing per dataset primo cloudword
daticw1=datiITA.loc[datiITA['timestamp'] >= '2020-08-23 21:22:00']
daticw1=daticw1.loc[daticw1['timestamp'] <= '2020-08-23 21:33:00']
daticw1
```

```python
#slicing dataset secondo cloudword
daticw2=datiITA.loc[datiITA['timestamp'] >= '2020-08-23 21:46:00']
daticw2=daticw2.loc[daticw2['timestamp'] <= '2020-08-23 21:57:00']
daticw2
```

```python
#slicing dataset terzo cloudword
daticw3=datiITA.loc[datiITA['timestamp'] >= '2020-08-23 22:17:00']
daticw3=daticw3.loc[daticw3['timestamp'] <= '2020-08-23 22:28:00']
daticw3
```

```python
daticw1.to_csv('cw1.csv',sep=';')
daticw2.to_csv('cw2.csv',sep=';')
daticw3.to_csv('cw3.csv',sep=';')
```
