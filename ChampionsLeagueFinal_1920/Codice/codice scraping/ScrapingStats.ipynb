```python
import requests
html_contents=requests.get('https://www.footballcritic.com/uefa-champions-league-paris-saint-germain-fc-fc-bayern-munchen/match-stats/2136879')
```

```python
!pip install -U beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

```python
html_soup=BeautifulSoup(html_contents.text,'html.parser')
```

```python
print(html_soup.prettify())
```

```python
My_table = html_soup.find('div',{'class':'stats-block'})
My_table
```

```python
td_data=My_table.find_all("span")
titoli=My_table.find_all("strong")
```

```python
main_list=[]
for element in range(0,len(td_data),3):
    main_list.append(td_data[element:element+3])
    
lista=[]
for titolo in range(0,len(titoli)):
    lista.append(titoli[titolo])
lista
```

```python
import pandas as pd
df2=pd.DataFrame(main_list,columns=['evento','PSG','Bayern Monaco'])
dati=df2.drop([0,1,2,3,22],axis=0)
dati=dati.reset_index(drop=True)
listaposs=[lista[0].text,lista[1].text]
del lista[0:2]
df2
```

```python
lista2=[]
for i in range(0,len(lista)):
    lista2.append(lista[i].text)
dati['evento']=lista2
```

```python
dati = dati.astype(str)
dati
```

```python
dati= dati.replace({'<span class="num home">':''}, regex=True).replace({'</span>':''},regex=True).replace({'<span class="num">':''},regex=True)
dati
```

```python
 dati.loc[-1] = ['Possession', listaposs[0],listaposs[1]] # adding a row
 dati.index = dati.index + 1  # shifting index
 dati = dati.sort_index()  # sorting by index
```
