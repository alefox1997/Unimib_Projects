```python
import time

hashtag="#OprahMeganHarry OR #MeganandHarryonOprah OR #meghanandharry"  


start_date ="2021-03-08"  #orario utc
end_date = "2021-03-09"

list_tweet=[]

backoff_counter = 1      
while True:
    try:
        for tweet in tweepy.Cursor(api.search,q=hashtag, since=start_date,until=end_date).items():
            try:
                    retweeted_status = tweet.retweeted_status
        
            except AttributeError:
                    retweeted_status = "None"
       
            list_tweet.append([tweet.created_at, tweet.id, tweet.id_str, tweet.text, retweeted_status, tweet.retweet_count, tweet.user.id_str,tweet.user.screen_name, tweet.user.location, tweet.lang, tweet.entities['hashtags'], tweet.entities['user_mentions'], tweet.coordinates]) 
        break
    except tweepy.TweepError as e:
        print(e.reason)
        time.sleep(60*backoff_counter)
        backoff_counter += 1
        continue
```

```python
tweets = pd.DataFrame(list_tweet[1:],columns=['created_at','id','id_str','text','retweeted_status','retweet_count','user.id_str','user.screen_name','user.location','lang','entities.hashtags','entities.user_mentions', 'coordinates'])
```
