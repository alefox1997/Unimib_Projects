```python
#import dati
import pandas as pd
import numpy as np
covid=pd.read_excel("C:/Users/aless/OneDrive/Desktop/AleBicocca/Data Science/Anno 2/Lab Biosciences/Project/Dati_Covid/covid_statistics.xlsx")
parenti=pd.read_excel("C:/Users/aless/OneDrive/Desktop/AleBicocca/Data Science/Anno 2/Lab Biosciences/Project/Dati_Covid/parenti_statistics.xlsx")
```

```python
covid=covid.drop("Unnamed: 0",axis=1)
parenti=parenti.drop("Unnamed: 0",axis=1)
```

```python
stand=[]
for index,row in covid[0:11].iterrows():
    val=round(row['Principal_Author']*1000/row['FTE_researchers'],3) #moltiplico per 100 per avere migliore risultato alla fine
    stand.append(val)
covid=covid[0:11]
covid['st_value_new']=stand
covid
#ordinamento per valore assoluto 
```

```python
stand=[]
for index,row in parenti[0:11].iterrows():
    val=round(row['Principal_Author']*1000/row['FTE_researchers_mean'],3) #moltiplico per 100 per avere migliore risultato alla fine
    stand.append(val)
parenti=parenti[0:11]
parenti['st_value_new']=stand
parenti
#ordinamento per valore assoluto 
```

```python
#chi è migliorato/peggiorato di più? chi ha confermato vantaggio?
```

```python
merge=pd.merge(covid[['Country','Country_code','Sum','Principal_Author','pct_sum','st_value','st_value_new']],parenti[['Country','Sum','Principal_Author','pct_sum','st_value','st_value_new']], on='Country')
merge.rename(columns={'Sum_x': 'Total_projects_covid', 'Sum_y': 'Total_projects_coronavirus','pct_sum_x':'pct_sum_cov','pct_sum_y':'pct_sum_corona','st_value_x':'st_value_cov','st_value_y':'st_value_coronavirus','st_value_new_x':'st_value_new_cov','st_value_new_y':'st_value_new_coronavirus','Principal_Author_x':'Principal_Author_cov','Principal_Author_y':'Principal_Author_coronavirus'}, inplace=True)
merge
```

```python
#chi è migliorato di più tra i paesi? calcoliamo un indicatore c he sfrutta i valori standardizzati
delta_pct=[]
delta_pct_eti=[]
indice_std=[]
indice=[]
indice_pa_std=[]
merge=merge[0:11]
for index,row in merge.iterrows():
    
    delta=round((row['pct_sum_cov'] - row['pct_sum_corona']),2)
    delta_pct.append(delta)
    
    delta_pct_eti.append(str(delta)+"%")
    ind=round((row['Total_projects_covid'] - row['Total_projects_coronavirus'])/row['Total_projects_covid'],2)
    indice.append(ind)
    
    ind_std=round((row['st_value_cov'] - row['st_value_coronavirus'])/row['st_value_cov'],2)
    indice_std.append(ind_std)
    
    ind_pa=round((row['st_value_new_cov'] - row['st_value_new_coronavirus'])/row['st_value_new_cov'],2)
    indice_pa_std.append(ind_pa)
    
merge['delta_pct']=delta_pct
merge['delta_pct_eti']=delta_pct_eti
merge['indice_std']=indice_std
merge['indice']=indice 
merge['indice_pa_std']=indice_pa_std
merge
```

```python
merge.to_excel('indicatori_new.xlsx')
```
