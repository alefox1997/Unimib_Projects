```python
standard=[]
count=0
for column in df_merged_fin:
    count+=1
    serie=df_merged_fin[column]
    if count>1:
        if str(serie[19]) != 'nan':
            ris=round(serie[19]+serie[19]*serie['75%'],2)
            standard.append(ris)
        else:
            if count==10:
                ris=round(serie[10]+serie[10]*serie['75%'],2)
                standard.append(ris)
            if count==9 or count==14 or count==16:
                ris=round(serie[17]+serie[17]*serie['75%'],2)
                standard.append(ris)
            if count==2 or count==13:
                ris=round(serie[18]+serie[18]*serie['75%'],2)
                standard.append(ris)
        
standard
    
```
