```python
import pandas as pd
# import del csv preprocessato 
interjuve=pd.read_csv("interjuve_treetag.csv")
interjuve
```

```python
from advertools import extract_emoji

# extract_emoji permette di ottenere le emoji contenute in un testo
def remove_emojis(text):
    emoji_summary = extract_emoji(text)
    emojis = emoji_summary['emoji_flat']
    # si salva la stringa escludendo le emoji trovate precedentemente
    no_emoji_text = text.translate({ord(x): '' for x in emojis})
    
    return no_emoji_text,emojis
```

```python
tweet_test = interjuve['clean_text'].iloc[12]
print(tweet_test)
print(remove_emojis(tweet_test))
```

```python
# si aggiunge al csv il testo senza emoji e la lista delle emoji

noemoji_text = []
emoji = []
for index,row in interjuve.iterrows():
    clean_tweet = remove_emojis(row['clean_text'])[0]
    noemoji_text.append(clean_tweet)
    emojis_tweet = remove_emojis(row['clean_text'])[1]
    emoji.append(emojis_tweet)
    
    
interjuve['noemoji_text'] = noemoji_text
interjuve['emoji'] = emoji

interjuve
```

```python
#export del CSV
saved = interjuve.to_csv('interjuve_treetag_emoji.csv', index = True)
```
